```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
conversionDF = pd.read_csv('conversionRates.csv')
```

```python
conversionDF.head()
```

```python
conversionDF.drop('Unnamed: 0', axis = 1,inplace = True)
```

```python
conversionDF[conversionDF['exchangeRate'] == conversionDF['exchangeRate'].max()]
```

The Bharaini Dinar has the greatest exchange rate where 1 USD is equal to 2.652053 BHD. 

```python
conversionDF[conversionDF['exchangeRate'] == conversionDF['exchangeRate'].min()]
```

The Iranian Rial has the lowest exchange rate where 1 USD is equal to 0.00003 IRR. 

```python
(conversionDF['exchangeRate'] > 1).sum()
```

```python
(conversionDF['exchangeRate'] < 1).sum()
```

There are 6 currencies that are stronger than the USD, and there are 78 currencies that are weaker than the USD. 

```python
fig = plt.figure(figsize=(10,2))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(conversionDF['originCountry'].tolist(),conversionDF['exchangeRate'].tolist())
ax1.set_title('USD Foreign Exchange Rates')
ax1.set_xlabel('Currencies')
ax1.set_ylabel('Exchange Rate')
```

```python
lowRate = conversionDF[conversionDF['exchangeRate']>1]
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.plot(lowRate['originCountry'].tolist(),lowRate['exchangeRate'].tolist())
ax2.set_title('Strong Currencies')
ax2.set_xlabel('Currencies')
ax2.set_ylabel('Exchange Rate')
```

```python
highRate = conversionDF[conversionDF['exchangeRate']<1]
highRate['exchangeRate'].mean()
```

```python
higherRate = conversionDF[conversionDF['exchangeRate'] < highRate['exchangeRate'].mean()]
higherRate.info()
```

It is interesting to note that there are 56 currencies that have an exchange rate lower than 0.1541264743589744.

```python
fig3 = plt.figure()
ax3 = fig3.add_axes([0,0,1,1])
ax3.plot(higherRate['originCountry'].tolist(),higherRate['exchangeRate'].tolist())
ax3.set_title('Very Weak Currencies')
ax3.set_xlabel('Currencies')
ax3.set_ylabel('Exchange Rate')
```
